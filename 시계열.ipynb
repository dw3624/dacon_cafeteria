# 1. 데이터 가져오기

```python
train_ln = pd.read_csv('./data/train_ln.csv',parse_dates=['ymd'])
train_dn = pd.read_csv('./data/train_dn.csv',parse_dates=['ymd'])
test_ln = pd.read_csv('./data/test_ln.csv',parse_dates=['ymd'])
test_dn = pd.read_csv('./data/test_dn.csv',parse_dates=['ymd'])
```

```python
print('train_ln.shape: ',train_ln.shape)
print('train_dn.shape: ',train_dn.shape)
print('test_ln.shape: ',test_ln.shape)
print('test_dn.shape: ',test_dn.shape)
```

```python
train_ln['ln_prop'] = train_ln.ln / train_ln.atoffice
train_dn['dn_prop'] = train_dn.dn / train_dn.atoffice
```

```python
train_ln['year']=train_ln.ymd.dt.strftime('%Y')
train_ln['month']=train_ln.ymd.dt.strftime('%m')
train_ln['day']=train_ln.ymd.dt.strftime('%d')
```

```python
train_dn[['dow','dn_prop']].groupby('dow').mean().\
sort_values('dn_prop',ascending=False)
```

```python
import statsmodels.api as sm
```

```python
plt.figure(figsize=(20,2))
train_ln.ln.plot()
```

```python
tl = pd.Series(train_ln.ln)
tl.index = train_ln.ymd
res = sm.tsa.seasonal_decompose(tl,period=60)
```

```python
plt.figure(figsize=(20,2))
plt.plot(res.seasonal)
plt.show()
```

```python
tl_month = tl.groupby(tl.index.month).mean()
plt.plot(tl_month)
```

```python
fig = plt.figure(figsize=(8, 8))

# 自己相関(ACF)のグラフ
ax1 = fig.add_subplot(211)
sm.graphics.tsa.plot_acf(tl, lags=60, ax=ax1) #ACF計算とグラフ自動作成

# 偏自己相関(PACF)のグラフ
ax2 = fig.add_subplot(212)
sm.graphics.tsa.plot_pacf(tl, lags=60, ax=ax2) #PACF計算とグラフ自動作成

plt.tight_layout() # グラフ間スキマ調整
```

```python
tl_diff = tl-tl.shift()
tl_diff.head()
```

```python
tl_diff = tl_diff.dropna()
len(tl_diff)
```

```python
fig = plt.figure(figsize=(12, 4))

# 原型列のグラフ
ax1 = fig.add_subplot(211)
plt.plot(tl)

# 階差系列のグラフ
ax2 = fig.add_subplot(212)
plt.plot(tl_diff)

plt.tight_layout() # グラフ間スキマ調整
```

```python
# 階差系列データでコレログラム作成
fig = plt.figure(figsize=(8, 8))

# 自己相関(ACF)のグラフ
ax1 = fig.add_subplot(211)
sm.graphics.tsa.plot_acf(tl_diff, lags=60, ax=ax1) #ACF計算とグラフ自動作成

# 偏自己相関(PACF)のグラフ
ax2 = fig.add_subplot(212)
sm.graphics.tsa.plot_pacf(tl_diff, lags=60, ax=ax2) #PACF計算とグラフ自動作成

plt.tight_layout() # グラフ間スキマ調整
```

```python
import warnings
warnings.filterwarnings('ignore') # 計算警告を非表示
```

```python
res_selection = sm.tsa.arma_order_select_ic(tl_diff, ic='aic', trend='nc')
res_selection
```

```python
sarimax = sm.tsa.SARIMAX(tl, 
                        order=(4, 1, 1),
                        seasonal_order=(1, 1, 1, 5),
                        enforce_stationarity = False,
                        enforce_invertibility = False
                        ).fit()
```

```python
sarimax_resid = sarimax.resid
```

```python
# モデル残差のコレログラム
fig = plt.figure(figsize=(8, 8))

# 自己相関(ACF)のグラフ
ax1 = fig.add_subplot(211)
sm.graphics.tsa.plot_acf(sarimax_resid, lags=40, ax=ax1) #ACF計算とグラフ自動作成

# 偏自己相関(PACF)のグラフ
ax2 = fig.add_subplot(212)
sm.graphics.tsa.plot_pacf(sarimax_resid, lags=40, ax=ax2) #PACF計算とグラフ自動作成

plt.tight_layout() # グラフ間スキマ調整
```

```python
sarimax_resid
```

```python
train_ln['new_ln'] = sarimax_resid.values
week_encoded = train_ln.groupby('dow').new_ln.mean()
```

```python
week_encoded
```

```python
train_ln['new_dow'] = train_ln['dow'].replace(week_encoded)
```

```python
train_ln['new_new_ln'] = train_ln['new_ln']-train_ln['new_dow']
```

```python
train_ln.columns
```

```python
cols = ['employees', 'dayoff', 'bustrip', 'overtime', 'remote',
       'atoffice', 
       'new_dow', 'bulgogi']
feature_x = cols
feature_y = ['new_ln']
data_x = train_ln[feature_x]
data_y = train_ln[feature_y]
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_reg = RandomForestRegressor(n_estimators=10000)
rf_reg = rf_reg.fit(data_x, data_y)

fti = rf_reg.feature_importances_

dic_arr = {'importance':fti, 'feature':feature_x}
pd.DataFrame(dic_arr).sort_values('importance', ascending=False).reset_index(drop=True).loc[:15, :]
```

```python
df = train_ln[feature_x+feature_y]
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,:-1],
                                                    df.iloc[:,-1],
                                                    test_size=0.3,
                                                    random_state=42)
rtree = RandomForestRegressor(n_jobs=-1, random_state=42)
rtree.fit(X_train, y_train)
y_pred = rtree.predict(X_test)
```

```python
print("훈련 세트 정확도: {:.3f}".format(rtree.score(X_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(rtree.score(X_test, y_test)))
print("MAE: {:.3f}".format(mean_absolute_error(y_test,y_pred)))
```
