```python
train = pd.read_csv('./data/train.csv')
```

```python
train.head(2)
```

```python
train.columns = ['ymd', 'day', 'num_total', 'num_vac', 'num_btrip', 'num_overtime',
       'num_athome', 'menu1', 'menu2', 'menu3', 'lunch', 'dinner']
train['year'] = train['ymd'].str[:4].astype('int')
train['month'] = train['ymd'].str[5:7].astype('int')
train['date'] = train['ymd'].str[8:].astype('int')
```

```python
train = train[['ymd','year','month','date','day',
              'num_total','num_vac','num_btrip','num_overtime','num_athome',
              'menu1', 'menu2', 'menu3', 'lunch', 'dinner']]
```

```python
train.menu1 = train.menu1.str.replace(' \(\S+\)','')
train.menu2 = train.menu2.str.replace(' \(\S+\)','')
train.menu3 = train.menu3.str.replace(' \(\S+\)','')
```
